```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mat_mul(A,B):
    if len(A[0])!=len(B):
        return ValueError("no of columns in A must be equal to the length of vector u")
    result=[]
    for i in range(len(A)):
        sum=0
        for j in range(len(B)):
            sum+=A[i][j]*B[j]
        result.append(sum)
    return result
```

```python
D=[[3,0],[0,6]]
D
x=[1,1]
e2=[0,1]
e1=[1,0]
```

```python
# transforming e1,e2 and x using the diagonal matrix D
De1=mat_mul(D,e1)
De2=mat_mul(D,e2)
Dx=mat_mul(D,x)
```

```python
def plot_graph(end1, end2, end3, end4, end5, end6, title, lim1, lim2):
    # Define starting and ending points for arrows
    start1 = start2 = start3 = start4 = start5 = start6 = (0, 0)

    # Create a figure and axis
    fig, ax = plt.subplots()

    # Plot arrows
    arrow1 = ax.arrow(*start1, end1[0]-start1[0], end1[1]-start1[1], head_width=0.3, head_length=0.3, fc='blue', ec='blue') #e1
    arrow2 = ax.arrow(*start2, end2[0]-start2[0], end2[1]-start2[1], head_width=0.3, head_length=0.3, fc='red', ec='red') #e2
    arrow3 = ax.arrow(*start3, end3[0]-start3[0], end3[1]-start3[1], head_width=0.3, head_length=0.3, fc='green', ec='green') #x
    arrow4 = ax.arrow(*start4, end4[0]-start4[0], end4[1]-start4[1], head_width=0.3, head_length=0.3, fc='blue', ec='blue') #D*e1
    arrow5 = ax.arrow(*start5, end5[0]-start5[0], end5[1]-start5[1], head_width=0.3, head_length=0.3, fc='red', ec='red') #D*e2
    arrow6 = ax.arrow(*start6, end6[0]-start6[0], end6[1]-start6[1], head_width=0.3, head_length=0.3, fc='green', ec='green') #D*x

    # Add labels near the arrow heads
    ax.annotate(f'e1({end1[0]}, {end1[1]})', xy=(end1[0], end1[1]), xytext=(end1[0], end1[1]+1), textcoords='offset points')
    ax.annotate(f'e2({end2[0]}, {end2[1]})', xy=(end2[0], end2[1]), xytext=(end2[0], end2[1]+1), textcoords='offset points')
    ax.annotate(f'x({end3[0]}, {end3[1]})', xy=(end3[0], end3[1]), xytext=(end3[0]+3, end3[1]+3), textcoords='offset points')
    ax.annotate(f'De1({end4[0]}, {end4[1]})', xy=(end4[0], end4[1]), xytext=(end4[0], end4[1]+1), textcoords='offset points')
    ax.annotate(f'De2({end5[0]}, {end5[1]})', xy=(end5[0], end5[1]), xytext=(end5[0], end5[1]+1), textcoords='offset points')
    ax.annotate(f'Dx({end6[0]}, {end6[1]})', xy=(end6[0], end6[1]), xytext=(end6[0], end6[1]+1), textcoords='offset points')

    # Set limits and labels
    #ax.set_xlim(-1, 10)
    #ax.set_ylim(-1, 10)
    ax.set_xlim(lim1, lim2)
    ax.set_ylim(lim1, lim2)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title(title)

    # Show plot
    plt.grid()
    plt.show()
```

```python
plot_graph(e1, e2, x, De1, De2, Dx, "Scaling", -1, 10)
```

```python
plot_graph(e1, e2, x, De1, De2, Dx, "Rotation", -3, 3)
```

Diagonal matrix changes the direction of any vectors except its eigenvectors([1,0] & [0,1]) but it does not maintain the degree of angle it rotates a vector, so its just stretching of various axis and cannot be consider as rotation whereas rotation matrix constantly maintains the rotating angle for a rotation matrix.
