```python
import numpy as np
import matplotlib.pyplot as plt

# Define matrix A
A = np.array([[2, 0],
              [0, 5]])

# Define vectors u and v
u = np.array([2, 3])
v = np.array([-2, -5])

# Calculate Au and Av
Au = np.dot(A, u)
Av = np.dot(A, v)

# Plot the vectors
plt.figure()
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='b', label='u')
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r', label='v')
plt.quiver(0, 0, Au[0], Au[1], angles='xy', scale_units='xy', scale=1, color='g', label='Au')
plt.quiver(0, 0, Av[0], Av[1], angles='xy', scale_units='xy', scale=1, color='m', label='Av')

# Set plot limits
plt.xlim(-10, 10)
plt.ylim(-10, 10)

# Add labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('Vector Transformation under Matrix A')
plt.axhline(0, color='k',linewidth=0.5)
plt.axvline(0, color='k',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()

# Show plot
plt.show()
```
