```python
# Function to calculate the determinant of a 2x2 matrix
```

```python
def determinant_2x2(matrix):
    return matrix[0][0] * matrix[1][1] - matrix[0][1] * matrix[1][0]
```

```python
# Function to calculate the determinant of a 3x3 matrix
```

```python
def determinant_3x3(matrix):
    a, b, c = matrix[0]
    d, e, f = matrix[1]
    g, h, i = matrix[2]
    return a * determinant_2x2([[e, f], [h, i]]) - b * determinant_2x2([[d, f], [g, i]]) + c * determinant_2x2([[d, e], [g, h]])
```

```python
# Function to calculate the inverse of a matrix
```

```python
def inverse(matrix):
    det = determinant_3x3(matrix)
    if det == 0:
        return None  # Matrix is singular, inverse doesn't exist
    else:
        a, b, c = matrix[0]
        d, e, f = matrix[1]
        g, h, i = matrix[2]
        adjugate = [[determinant_2x2([[e, f], [h, i]]), -determinant_2x2([[b, c], [h, i]]), determinant_2x2([[b, c], [e, f]])],
                    [-determinant_2x2([[d, f], [g, i]]), determinant_2x2([[a, c], [g, i]]), -determinant_2x2([[a, c], [d, f]])],
                    [determinant_2x2([[d, e], [g, h]]), -determinant_2x2([[a, b], [g, h]]), determinant_2x2([[a, b], [d, e]])]]
        inverse_matrix = [[adjugate[i][j] / det for j in range(3)] for i in range(3)]
        return inverse_matrix
```

```python
# Calculate the inverse of matrix A    
```

```python
inverse_A = inverse(A)
```

```python
# Print the inverse of matrix A
```

```python
print("Inverse of matrix A:")
for row in inverse_A:
    print(row)
```

```python
# Function to multiply two matrices
```

```python
def multiply_matrices(matrix1, matrix2):
    result = [[0, 0, 0] for _ in range(3)]
    for i in range(3):
        for j in range(3):
            for k in range(3):
                result[i][j] += matrix1[i][k] * matrix2[k][j]
    return result
```

```python
# Calculate the product of A and its inverse
```

```python
product = multiply_matrices(A, inverse_A)
```

```python
# Print the product
```

```python
print("\nProduct of A and its inverse:")
for row in product:
    print(row)
```
