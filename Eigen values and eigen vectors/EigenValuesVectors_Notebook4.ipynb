**Step 1: Transition Matrix Generation**
# Define the page-link structure

```python
import numpy as np


page_links = {
    1: [2, 3, 4],
    2: [1, 3],
    3: [1, 4],
    4: [1]
}
```

# Total number of pages

```python
num_pages = len(page_links)
```

# Initialize the transition matrix with zeros

```python
transition_matrix = np.zeros((num_pages, num_pages))
```

# Fill the transition matrix based on the page-link structure

```python

for i in range(1, num_pages + 1):
    linked_pages = page_links[i]
    num_linked_pages = len(linked_pages)
    for j in linked_pages:
        transition_matrix[j - 1, i - 1] = 1 / num_linked_pages
```

**Step 2: Eigenvalue and Eigenvector Calculation**
# Calculate eigenvalues and eigenvectors

```python
eigenvalues, eigenvectors = np.linalg.eig(transition_matrix)
```

# Find the index of the eigenvalue closest to 1

```python
closest_index = np.argmin(np.abs(eigenvalues - 1))
closest_eigenvector = eigenvectors[:, closest_index]
```

**Step 3: Scaling Eigenvector**
# Rescale the eigenvector to ensure the sum of all entries equals 1

```python
page_ranks = closest_eigenvector / np.sum(closest_eigenvector)
```

**Step 4: Page Importance Display**
# Display the importance of each page (PageRank)

```python
print("Page Importance (PageRank):")
for i, rank in enumerate(page_ranks, start=1):
    print(f"Page {i}: {rank}")
```

**Step 5: Page Ranking**
# Rank the pages based on their PageRank scores in descending order

```python
ranked_pages = np.argsort(page_ranks)[::-1] + 1  # Add 1 to get page numbers
print("\nPage Ranking (Descending Order):")
for rank, page in enumerate(ranked_pages, start=1):
    print(f"Rank {rank}: Page {page}")
```
