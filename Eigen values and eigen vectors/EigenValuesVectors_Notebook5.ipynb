1. Define the Transition Matrix:
 The transition matrix represents the probabilities of moving from one state (or page in this case) to another. Each row of the matrix corresponds to the probabilities of transitioning from the current state (page) to all possible next states (pages).
2. Initialize the Initial State Vector (𝑥): 
 This vector represents the initial probabilities of starting from each page. For example, if we start from Page 1, the initial state vector might be [1, 0, 0, 0] indicating that we start with 100% probability of being on Page 1.
3. Perform Matrix-Vector Multiplication:
 Multiply the transition matrix by the initial state vector. This multiplication calculates the probabilities of being on each page after one transition.
4. Repeat Multiplication:
 Repeat the matrix-vector multiplication process multiple times (in this case, 100 times). Each iteration represents a transition from one page to another.
5. Steady-State Distribution (𝑥):
 After repeating the multiplication process for a large number of iterations, the resulting vector x represents the steady-state distribution. It shows the long-term probabilities of being on each page regardless of the initial page.
6. Interpretation: Analyze the resulting vector x  :
 Each element of x represents the proportion of time spent on the corresponding page in the long run. This allows us to understand the relative importance or popularity of each page in the system.

**Define the transition matrix**

```python
import numpy as np


transition_matrix = np.array([
    [0, 0, 1, 0.5],
    [1/3, 0, 0, 0],
    [1/3, 0.5, 0, 0.5],
    [1/3, 0.5, 0, 0]
])
```

**Define the initial state vector**

```python
x_1 = np.array([1, 0, 0, 0])  # Starting from Page 1
x_3 = np.array([0, 0, 1, 0])  # Starting from Page 3
```

**Perform matrix-vector multiplication to simulate transitions**

```python
def simulate_transitions(transition_matrix, x, num_iterations):
    for _ in range(num_iterations):
        x = np.dot(x, transition_matrix)
    return x
```

**Simulate transitions starting from Page 1**

```python
num_iterations = 100
result_x_1 = simulate_transitions(transition_matrix, x_1, num_iterations)
print("Composition of vector x after", num_iterations, "iterations starting from Page 1:")
print(result_x_1)
```

**Simulate transitions starting from Page 3**

```python

result_x_3 = simulate_transitions(transition_matrix, x_3, num_iterations)
print("\nComposition of vector x after", num_iterations, "iterations starting from Page 3:")
print(result_x_3)
```

**what is this x called??**
